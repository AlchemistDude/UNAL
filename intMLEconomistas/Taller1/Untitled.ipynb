# Zoo animals

This dataset contains animal data from a zoo, the columns are the following:
- animal_name: Name of the animal.
- hair: it has hair or not.
- feathers: it has feathers or not.
- eggs: it has eggs or not.
- milk: it produces milk or not.
- airbone: it has airbone or not.
- aquatic: it is aquatic or not.
- predator: it is a predator or not.
- toothed: it has tooths or not.
- backbone: it has backbone or not.
- breathes: it breathes or not.
- venomous: it is venomous or not.
- fins: it has fins or not.
- legs: it has legs or not.
- tail: it has tail or not.
- domestic: it is domestic or not.
- catsize: it it catsize or not.
- classtype: 1 for Mammal,2 for Bird, 3 for Reptile, 4 for Fish, 5 for Amphibian, 6 for Bug and 7 for invertebrate.
